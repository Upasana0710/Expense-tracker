```python
# Generate Dummy User data

!pip install pandas
!pip install openpyxl
import pandas as pd
import random

# Function to generate random user data
def generate_user_data():
    userid = ["user" + str(i) for i in range(1, 101)]
    names = ["User" + str(i) for i in range(1, 101)]
    emails = [f"user{i}@example.com" for i in range(1, 101)]
    ages = [random.randint(18, 60) for _ in range(100)]
    financial_knowledge = random.choices(["BEGINNER", "INTERMEDIATE", "ADVANCED"], k=100)
    monthly_incomes = [random.randint(20000, 200000) for _ in range(100)]

    # Define states
    states = ["Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh",
              "Goa", "Gujarat", "Haryana", "Himachal Pradesh", "Jharkhand",
              "Karnataka", "Kerala", "Madhya Pradesh", "Maharashtra", "Manipur",
              "Meghalaya", "Mizoram", "Nagaland", "Odisha", "Punjab", "Rajasthan",
              "Sikkim", "Tamil Nadu", "Telangana", "Tripura", "Uttar Pradesh",
              "Uttarakhand", "West Bengal"]

    # Assign states randomly
    locations = random.choices(states, k=100)

    risk_tolerances = random.choices(["CONSERVATIVE", "MODERATE", "AGGRESSIVE"], k=100)

    return {
        "id": userid,
        "name": names,
        "email": emails,
        "age": ages,
        "financialKnowledge": financial_knowledge,
        "monthlyIncome": monthly_incomes,
        "location": locations,
        "riskTolerance": risk_tolerances,
    }

# Generate user data
user_data = generate_user_data()

# Create a DataFrame from the user data
df = pd.DataFrame(user_data)

# Export the DataFrame to an Excel sheet
df.to_excel("user_data.xlsx", index=False)
```

```python
# Generate Dummy expeces of user
!pip install faker

import pandas as pd
import random
from faker import Faker

fake = Faker()

# Load user data
user_data = pd.read_excel("user_data.xlsx")

# Function to generate random expenses data
def generate_expenses_data(users, num_expenses=1000):
    expenses_data = {
        "userID": random.choices(users["id"], k=num_expenses),
        "transactionDate": [fake.date_this_year() for _ in range(num_expenses)],
        "category": random.choices(
            [
                "Groceries",
                "Utilities",
                "Rent/Mortgage",
                "Transportation",
                "Healthcare",
                "Insurance",
                "Entertainment",
                "Dining Out",
                "Shopping",
                "Travel",
                "Education",
                "Savings/Investments",
                "Debt Payments",
                "Other",
            ],
            k=num_expenses,
        ),
        "amount": [round(random.uniform(10, 200000), 2) for _ in range(num_expenses)],
        "description": [fake.sentence() for _ in range(num_expenses)],
        "paymentMethod": random.choices(["Credit Card", "Debit Card", "Cash"], k=num_expenses),
    }

    return expenses_data

# Generate expenses data
expenses_data = generate_expenses_data(user_data)

# Create a DataFrame from the expenses data
expenses_df = pd.DataFrame(expenses_data)

# Export the DataFrame to an Excel sheet
expenses_df.to_excel("expenses_data.xlsx", index=False)
```

```python
expence_data = pd.read_excel("expenses_data.xlsx")
user_data = pd.read_excel("user_data.xlsx")
print(expence_data)
print(user_data)
```

```python
# Combining the data
final_data = pd.merge(expence_data, user_data, left_on='userID', right_on='id', suffixes=('_expense', '_user'))
final_data['transactionDate'] = pd.to_datetime(final_data['transactionDate'])
final_data['transactionDate'] = final_data['transactionDate'].astype(int)
final_data=final_data.drop(columns=['description', 'paymentMethod', 'id'])

# Impute missing values for numerical columns with the mean
final_data['age'] = final_data['age'].fillna(final_data['age'].mean())
final_data['monthlyIncome'] = final_data['monthlyIncome'].fillna(final_data['monthlyIncome'].mean())

# Impute missing values for categorical columns with the mode
final_data['location'] = final_data['location'].fillna(final_data['location'].mode()[0])
final_data['riskTolerance'] = final_data['riskTolerance'].fillna(final_data['riskTolerance'].mode()[0])

final_data.to_excel("final_data.xlsx", index=False)
```
